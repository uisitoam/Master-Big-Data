```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import Ridge, Lasso, LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, classification_report, roc_auc_score

# Preparamos los datos para los modelos de selección de características
# Primero separamos las variables predictoras y la variable objetivo
X = df.drop(['customerID', 'Churn'], axis=1, errors='ignore')
y = df['Churn'].map({'Yes': 1, 'No': 0})

# Identificamos columnas numéricas y categóricas
numeric_features = X.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_features = X.select_dtypes(include=['object']).columns.tolist()

# Configuramos el preprocesamiento
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_features),
        ('cat', OneHotEncoder(drop='first', sparse_output=False), categorical_features)
    ])

# Aplicamos el preprocesamiento y dividimos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Aplicamos la transformación
X_train_preprocessed = preprocessor.fit_transform(X_train)
X_test_preprocessed = preprocessor.transform(X_test)

# Obtenemos los nombres de las características después del preprocesamiento
ohe_feature_names = preprocessor.named_transformers_['cat'].get_feature_names_out(categorical_features)
all_feature_names = np.concatenate([numeric_features, ohe_feature_names])

print(f"Número de características después del preprocesamiento: {len(all_feature_names)}")

# 1. SELECCIÓN DE CARACTERÍSTICAS CON RIDGE
print("\n\n--- SELECCIÓN DE CARACTERÍSTICAS CON RIDGE ---")

# Utilizamos LogisticRegression con regularización Ridge (l2)
alphas = [0.001, 0.01, 0.1, 1, 10, 100]
ridge_model = LogisticRegression(penalty='l2', solver='liblinear', max_iter=1000)

param_grid = {'C': [1/alpha for alpha in alphas]}  # C es inverso de alpha
grid_search_ridge = GridSearchCV(ridge_model, param_grid, cv=5, scoring='accuracy')
grid_search_ridge.fit(X_train_preprocessed, y_train)

best_ridge = grid_search_ridge.best_estimator_
print(f"Mejor parámetro C: {grid_search_ridge.best_params_['C']}")

# Evaluamos el modelo Ridge
y_pred_ridge = best_ridge.predict(X_test_preprocessed)
print(f"Precisión del modelo Ridge: {accuracy_score(y_test, y_pred_ridge):.4f}")

# Obtenemos la importancia de las características
ridge_importance = pd.DataFrame({
    'Feature': all_feature_names,
    'Importance': np.abs(best_ridge.coef_[0])
})
ridge_importance = ridge_importance.sort_values('Importance', ascending=False)

print("\nTop 15 características más relevantes según Ridge:")
print(ridge_importance.head(15))

# Visualización
plt.figure(figsize=(12, 8))
sns.barplot(x='Importance', y='Feature', data=ridge_importance.head(15))
plt.title('Top 15 características según Ridge')
plt.tight_layout()
plt.show()

# 2. SELECCIÓN DE CARACTERÍSTICAS CON LASSO
print("\n\n--- SELECCIÓN DE CARACTERÍSTICAS CON LASSO ---")

# Utilizamos LogisticRegression con regularización Lasso (l1)
lasso_model = LogisticRegression(penalty='l1', solver='liblinear', max_iter=1000)

param_grid = {'C': [1/alpha for alpha in alphas]}
grid_search_lasso = GridSearchCV(lasso_model, param_grid, cv=5, scoring='accuracy')
grid_search_lasso.fit(X_train_preprocessed, y_train)

best_lasso = grid_search_lasso.best_estimator_
print(f"Mejor parámetro C: {grid_search_lasso.best_params_['C']}")

# Evaluamos el modelo Lasso
y_pred_lasso = best_lasso.predict(X_test_preprocessed)
print(f"Precisión del modelo Lasso: {accuracy_score(y_test, y_pred_lasso):.4f}")

# Obtenemos la importancia de las características
lasso_importance = pd.DataFrame({
    'Feature': all_feature_names,
    'Importance': np.abs(best_lasso.coef_[0])
})
lasso_importance = lasso_importance.sort_values('Importance', ascending=False)

print("\nTop 15 características más relevantes según Lasso:")
print(lasso_importance.head(15))

# Características seleccionadas (con coeficientes no nulos)
selected_by_lasso = lasso_importance[lasso_importance['Importance'] > 0]
print(f"\nNúmero de características seleccionadas por Lasso: {len(selected_by_lasso)}")

# Visualización
plt.figure(figsize=(12, 8))
sns.barplot(x='Importance', y='Feature', data=lasso_importance.head(15))
plt.title('Top 15 características según Lasso')
plt.tight_layout()
plt.show()

# 3. SELECCIÓN DE CARACTERÍSTICAS CON RANDOM FOREST
print("\n\n--- SELECCIÓN DE CARACTERÍSTICAS CON RANDOM FOREST ---")

# Entrenamos un Random Forest Classifier
rf_model = RandomForestClassifier(random_state=42, n_estimators=100)
rf_model.fit(X_train_preprocessed, y_train)

# Evaluamos el modelo Random Forest
y_pred_rf = rf_model.predict(X_test_preprocessed)
print(f"Precisión del modelo Random Forest: {accuracy_score(y_test, y_pred_rf):.4f}")

# Obtenemos la importancia de las características
rf_importance = pd.DataFrame({
    'Feature': all_feature_names,
    'Importance': rf_model.feature_importances_
})
rf_importance = rf_importance.sort_values('Importance', ascending=False)

print("\nTop 15 características más relevantes según Random Forest:")
print(rf_importance.head(15))

# Visualización
plt.figure(figsize=(12, 8))
sns.barplot(x='Importance', y='Feature', data=rf_importance.head(15))
plt.title('Top 15 características según Random Forest')
plt.tight_layout()
plt.show()

# Comparación de los tres métodos
print("\n\n--- COMPARACIÓN DE LOS TRES MÉTODOS DE SELECCIÓN ---")

# Tomamos las top 10 características de cada método
top_ridge = set(ridge_importance.head(10)['Feature'])
top_lasso = set(lasso_importance.head(10)['Feature'])
top_rf = set(rf_importance.head(10)['Feature'])

# Características que aparecen en los tres métodos
common_features = top_ridge.intersection(top_lasso).intersection(top_rf)
print(f"\nCaracterísticas que aparecen en los tres métodos: {common_features}")

# Características que aparecen en al menos dos métodos
in_two_methods = (top_ridge.intersection(top_lasso)).union(
                  top_ridge.intersection(top_rf)).union(
                  top_lasso.intersection(top_rf)) - common_features
print(f"\nCaracterísticas que aparecen en al menos dos métodos: {in_two_methods}")

# Análisis de consenso: suma de rangos normalizados
combined_ranks = pd.DataFrame({'Feature': all_feature_names})

# Normalizamos y añadimos los rangos de cada método
for name, imp_df in [('Ridge', ridge_importance), ('Lasso', lasso_importance), ('RandomForest', rf_importance)]:
    # Creamos un diccionario de características a rangos
    ranks = {feat: idx + 1 for idx, feat in enumerate(imp_df['Feature'])}
    # Añadimos los rangos al dataframe combinado
    combined_ranks[f'{name}_rank'] = combined_ranks['Feature'].map(ranks).fillna(len(all_feature_names))
    # Normalizamos los rangos (0-1)
    combined_ranks[f'{name}_norm'] = combined_ranks[f'{name}_rank'] / len(all_feature_names)

# Calculamos la puntuación consenso (suma de rangos normalizados)
combined_ranks['consensus_score'] = (3 - (combined_ranks['Ridge_norm'] + 
                                         combined_ranks['Lasso_norm'] + 
                                         combined_ranks['RandomForest_norm']))

# Ordenamos por la puntuación consenso
combined_ranks = combined_ranks.sort_values('consensus_score', ascending=False)

print("\nTop 15 características según el consenso de los tres métodos:")
print(combined_ranks[['Feature', 'consensus_score']].head(15))

# Visualización del consenso
plt.figure(figsize=(12, 8))
sns.barplot(x='consensus_score', y='Feature', data=combined_ranks.head(15))
plt.title('Top 15 características según el consenso de los tres métodos')
plt.tight_layout()
plt.show()

# Resumen: las mejores características para predecir la deserción de clientes
print("\n\n--- RESUMEN DE LAS CARACTERÍSTICAS MÁS RELEVANTES ---")
print("Las características más importantes para predecir la deserción de clientes son:")
for idx, row in combined_ranks.head(10).iterrows():
    print(f"- {row['Feature']}: {row['consensus_score']:.4f}")
```

```python
# SELECCIÓN DE CARACTERÍSTICAS CON MÚLTIPLES MODELOS DE APRENDIZAJE SUPERVISADO

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import roc_auc_score
from sklearn.feature_selection import SelectFromModel

# Modelos para selección de características
from sklearn.linear_model import Ridge, Lasso, LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier

import pandas as pd

df = pd.read_csv("Telco-Customer-Churn.csv")
df = pd.read_csv("Telco-Customer-Churn.csv")

# Ver información detallada
print("Forma original del dataframe:", df.shape)
print("\nValores únicos en columna Churn:")
print(df['Churn'].value_counts())
print("\nInformación de columnas:")
print(df.dtypes)

# Convertir TotalCharges a numérico y manejar valores nulos
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df['TotalCharges'].fillna(df['MonthlyCharges'], inplace=True)

# Verificar valor nulo potencial en Churn
print("\nValores nulos por columna después de convertir TotalCharges:")
print(df.isnull().sum())

# Función mejorada para preparar datos
def prepare_data_for_supervised(df):
    # Copiar el dataframe para no modificar el original
    data = df.copy()
    # Eliminar filas con valores nulos en cualquier columna
    data_clean = data.dropna()
    
    if len(data_clean) < len(data):
        print(f"Se eliminaron {len(data) - len(data_clean)} filas con valores nulos")
    
    # Codificar la variable objetivo explícitamente
    data_clean['Churn_Encoded'] = data_clean['Churn'].map({'Yes': 1, 'No': 0})  
    
    # Si aún hay nulos, eliminar esas filas
    if data_clean['Churn_Encoded'].isnull().sum() > 0:
        data_clean = data_clean.dropna(subset=['Churn_Encoded'])
        print(f"Se eliminaron {len(data_clean) - len(data)} filas adicionales con Churn nulo")
    
    # Convertir variables categóricas a numéricas
    le = LabelEncoder()
    categorical_cols = []
    
    for col in data_clean.columns:
        if data_clean[col].dtype == 'object' and col != 'customerID':
            data_clean[col] = le.fit_transform(data_clean[col])
            categorical_cols.append(col)
    
    # Eliminar customerID si existe
    if 'customerID' in data_clean.columns:
        data_clean = data_clean.drop('customerID', axis=1)
    
    print(f"Dimensiones del dataframe después de preprocesamiento: {data_clean.shape}")
    
    return data_clean, categorical_cols


# Preparar los datos
processed_data, categorical_cols = prepare_data_for_supervised(df)

# Verificación final
print("\nVerificación final:")
print("Forma del dataframe procesado:", processed_data.shape)
print("Valores nulos en dataframe procesado:")
print(processed_data.isnull().sum().sum())

# Separar features y target
X = processed_data.drop(['Churn', 'Churn_Encoded'], axis=1)
y = processed_data['Churn_Encoded']

# Dividir en conjunto de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

print("\nForma de X_train:", X_train.shape)
print("Forma de X_test:", X_test.shape)

# Escalar las características
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python

# 2. Entrenamiento de múltiples modelos y selección de características
models = {
    'Ridge': Ridge(alpha=1.0, random_state=42),
    'Lasso': Lasso(alpha=0.01, random_state=42),
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1),
    'XGBoost': XGBClassifier(n_estimators=100, random_state=42, n_jobs=-1, learning_rate=0.1),
    'SVM': SVC(kernel='linear', probability=True, random_state=42)
}

feature_importance_df = pd.DataFrame(index=X.columns)

# Entrenar modelos y registrar importancia de características
for name, model in models.items():
    print(f"\nEntrenando modelo: {name}")
    
    # Entrenar modelo
    if name in ['Ridge', 'Lasso']:
        # Para modelos lineales usamos LogisticRegression para clasificación
        clf = LogisticRegression(penalty='l2' if name=='Ridge' else 'l1', solver='saga', random_state=42, 
                                C=1.0 if name=='Ridge' else 0.1, max_iter=1000)
        clf.fit(X_train_scaled, y_train)
        
        # Obtener coeficientes para importancia
        importance = np.abs(clf.coef_[0])
        
    elif name == 'SVM':
        # Para SVM lineal
        clf = models[name]
        clf.fit(X_train_scaled, y_train)
        
        # Obtener coeficientes para importancia en SVM lineal
        importance = np.abs(clf.coef_[0])
        
    else:
        # Para RF y XGBoost
        clf = models[name]
        clf.fit(X_train_scaled, y_train)
        
        # Obtener importancia
        importance = clf.feature_importances_
    
    # Evaluar rendimiento
    if name in ['Ridge', 'Lasso', 'SVM']:
        y_pred_proba = clf.predict_proba(X_test_scaled)[:, 1]
    else:
        y_pred_proba = clf.predict_proba(X_test_scaled)[:, 1]
    
    auc = roc_auc_score(y_test, y_pred_proba)
    print(f"AUC en conjunto de prueba: {auc:.4f}")
    
    # Guardar importancia de características
    feature_importance_df[name] = importance

# 3. Agregar el ranking de importancia para cada modelo
for col in feature_importance_df.columns:
    feature_importance_df[f"{col}_rank"] = feature_importance_df[col].rank(ascending=False)

# Calcular ranking promedio entre todos los modelos
rank_columns = [col for col in feature_importance_df.columns if col.endswith('_rank')]
feature_importance_df['avg_rank'] = feature_importance_df[rank_columns].mean(axis=1)

# Ordenar por ranking promedio
feature_importance_df = feature_importance_df.sort_values('avg_rank')

# Mostrar top 15 características
print("\n--- TOP 15 CARACTERÍSTICAS MÁS IMPORTANTES ---")
print(feature_importance_df[['avg_rank'] + feature_importance_df.columns[:-1].tolist()].head(15))

# 4. Visualizar importancia de características en un solo gráfico
top_features = feature_importance_df.head(15).index

# Crear un único dataframe para la visualización
viz_df = pd.DataFrame()
for model in models.keys():
    temp_df = pd.DataFrame({
        'Feature': top_features,
        'Importance': feature_importance_df.loc[top_features, model],
        'Model': model
    })
    viz_df = pd.concat([viz_df, temp_df])

# Normalizar la importancia dentro de cada modelo para mejor comparación
for model in models.keys():
    model_max = viz_df[viz_df['Model'] == model]['Importance'].max()
    viz_df.loc[viz_df['Model'] == model, 'Importance'] = viz_df.loc[viz_df['Model'] == model, 'Importance'] / model_max

plt.figure(figsize=(14, 10))
sns.barplot(x='Importance', y='Feature', hue='Model', data=viz_df)
plt.title('Top 15 Características Más Importantes por Modelo')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python

# 5. Seleccionar top N características para clustering
N = 5  # Número de características a seleccionar
selected_features = feature_importance_df.index[:N].tolist()
print(f"\nCaracterísticas seleccionadas para clustering: {selected_features}")

# 6. Realizar clustering con características seleccionadas
print("\nREALIZANDO CLUSTERING CON CARACTERÍSTICAS SELECCIONADAS")

# Función para preparar datos de clustering
def prepare_for_clustering(df, target_value, selected_features):
    # Filtrar por target
    filtered_df = df[df['Churn'] == target_value].copy()
    
    # Obtener características numéricas y categóricas
    num_features = []
    cat_features = []
    
    for feature in selected_features:
        if feature in df.columns:
            if df[feature].dtype in ['int64', 'float64']:
                num_features.append(feature)
            else:
                cat_features.append(feature)
    
    # Crear dummies para categóricas
    dummy_cols = []
    if cat_features:
        for feature in cat_features:
            dummy = pd.get_dummies(filtered_df[feature], prefix=feature)
            filtered_df = pd.concat([filtered_df, dummy], axis=1)
            dummy_cols.extend(dummy.columns.tolist())
    
    # Seleccionar características finales
    X_cluster = filtered_df[num_features + dummy_cols].copy()
    X_cluster = X_cluster.dropna()
    
    # Escalar datos
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X_cluster)
    
    return filtered_df, X_cluster, X_scaled

# Preparar datos para clustering
churners, X_churners, X_churners_scaled = prepare_for_clustering(df, 'Yes', selected_features)
non_churners, X_non_churners, X_non_churners_scaled = prepare_for_clustering(df, 'No', selected_features)

# 7. Aplicar K-Means para ambos grupos
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

# Determinar número óptimo de clusters
n_clusters = 3  # Podemos usar el método del codo o silhouette para determinar esto

# Aplicar K-Means para desertores
kmeans_churners = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
churners_clusters = kmeans_churners.fit_predict(X_churners_scaled)
churners.loc[:, 'Cluster'] = churners_clusters

# Aplicar K-Means para no desertores
kmeans_non_churners = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
non_churners_clusters = kmeans_non_churners.fit_predict(X_non_churners_scaled)
non_churners.loc[:, 'Cluster'] = non_churners_clusters

# 8. Visualizar clusters con PCA
# Reducir dimensionalidad para visualización
pca = PCA(n_components=2)

# PCA para desertores
X_churners_pca = pca.fit_transform(X_churners_scaled)
explained_variance_churners = pca.explained_variance_ratio_

# PCA para no desertores
pca = PCA(n_components=2)
X_non_churners_pca = pca.fit_transform(X_non_churners_scaled)
explained_variance_non_churners = pca.explained_variance_ratio_

# Crear dataframes para visualización
churners_viz = pd.DataFrame({
    'x': X_churners_pca[:, 0],
    'y': X_churners_pca[:, 1],
    'Cluster': churners_clusters
})

non_churners_viz = pd.DataFrame({
    'x': X_non_churners_pca[:, 0],
    'y': X_non_churners_pca[:, 1],
    'Cluster': non_churners_clusters
})
```

```python

# 11. Visualizar características principales por cluster
def plot_cluster_features(data, feature, target, title=None):
    plt.figure(figsize=(14, 6))
    sns.boxplot(x='Cluster', y=feature, data=data)
    plt.title(title or f"{feature} por Cluster ({target})")
    plt.tight_layout()
    plt.show()

# Visualizar algunas características importantes
for feature in ['tenure', 'MonthlyCharges', 'Contract', 'TotalCharges', 'OnlineSecurity']:
    plot_cluster_features(churners, feature, "Desertores")
    plot_cluster_features(non_churners, feature, "No Desertores")

# Visualizar distribución de contrato por cluster
plt.figure(figsize=(16, 6))

# Para desertores
plt.subplot(1, 2, 1)
contract_by_cluster = pd.crosstab(churners['Cluster'], churners['Contract'], normalize='index')
contract_by_cluster.plot(kind='bar', stacked=True, ax=plt.gca())
plt.title('Tipo de Contrato por Cluster (Desertores)')
plt.xlabel('Cluster')
plt.ylabel('Proporción')

# Para no desertores
plt.subplot(1, 2, 2)
contract_by_cluster = pd.crosstab(non_churners['Cluster'], non_churners['Contract'], normalize='index')
contract_by_cluster.plot(kind='bar', stacked=True, ax=plt.gca())
plt.title('Tipo de Contrato por Cluster (No Desertores)')
plt.xlabel('Cluster')
plt.ylabel('Proporción')

plt.tight_layout()
plt.show()
```
