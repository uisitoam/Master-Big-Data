### 1. Que series recomendarías ver a alguen que unicamente viu Lost? 
- Ozark
- Sex Education

```python
datos = list(df["series"].apply(lambda x: x.split(",")))
from mlxtend.preprocessing import TransactionEncoder

encoder = TransactionEncoder()
datos_codificados = encoder.fit_transform(datos)
df = pd.DataFrame(datos_codificados, columns=encoder.columns_)

df
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
frequent_itemsets = apriori(df, min_support=0.01, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.1)
lost_recommendations = rules[rules['antecedents'].apply(lambda x: 'Lost' in x)]

lost_recommendations = lost_recommendations.sort_values('confidence', ascending=False)

for i, row in lost_recommendations.iterrows():
    consequents = list(row['consequents'])
    print(f"- {consequents[0]} (Confianza: {row['confidence']:.2f}, Soporte: {row['support']:.2f})")
```

### 2. Cal é a serie máis vista?

```python
series_ordenadas = frequent_itemsets.sort_values('support', ascending=False)
series_ordenadas
```

### 3. Cal é a probabilidade de que alguen vira Two and a half men?

```python
p2men = df["Two and a half men"].sum() / len(df)
print(f"Probabilidad de ver Two and a half men: {p2men:.4f}")
```

### 4. Que probabildade hai de que alguen vira The Mentalist ou Chernobyl?

```python
pment = df["The Mentalist"].sum() / len(df)
pcher = df["Chernobyl"].sum() / len(df)

#pmentcher = (df["The Mentalist"] & df["Chernobyl"]).sum() / len(df)
#pment_o_cher = pment + pcher - pmentcher
pment_o_cher = pment + pcher

print(f"Probabilidad de ver The Mentalist o Chernobyl: {pment_o_cher:.4f}")
```

### 5. Con que probabilidade alguén que veu Lucifer verá Ozark?

```python
lucifer_to_ozark = rules[(rules['antecedents'].apply(lambda x: 'Lucifer' in x and len(x) == 1)) & 
                        (rules['consequents'].apply(lambda x: 'Ozark' in x and len(x) == 1))]

print(f"Confianza: {lucifer_to_ozark.iloc[0]['confidence']:.4f}")
```
